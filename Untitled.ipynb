```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def apply_filter(image, filter):
    
    return cv2.filter2D(image, -1, filter)

def main(image_path):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    
    first = np.array([[-1, -1, -1], [2, 2, 2], [-1, -1, -1]])
    second = np.array([[2, -1, -1], [-1, 2, -1], [-1, -1, 2]])
    third = np.array([[-1, 2, -1], [-1, 2, -1], [-1, 2, -1]])
    fourth = np.array([[-1, -1, 2], [-1, 2, -1], [2, -1, -1]])

   
    result1 = apply_filter(image, first)
    result2 = apply_filter(image, second)
    result3 = apply_filter(image, third)
    result4 = apply_filter(image, fourth)

   
    
    plt.figure(figsize=(10, 10))
    plt.subplot(2, 2, 1), plt.imshow(result1, cmap='gray'), plt.title('pic1')
    plt.subplot(2, 2, 2), plt.imshow(result2, cmap='gray'), plt.title('pic2')
    plt.subplot(2, 2, 3), plt.imshow(result3, cmap='gray'), plt.title('pic3')
    plt.subplot(2, 2, 4), plt.imshow(result4, cmap='gray'), plt.title('final pic')
    
    plt.show()


main('tiger.jpg')
```
